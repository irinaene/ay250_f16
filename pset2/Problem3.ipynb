```python
import numpy as np
import matplotlib.pyplot as plt
import corner

from scipy.optimize import least_squares
import emcee

% matplotlib inline
```

## Part (a) -- fit a line to a histogram!

```python
### Salpeter's data
lnM = np.array([2.23, 2.08, 1.93, 1.78, 1.63, 1.48, 1.33, 1.20, 1.09, 1.00, 0.93, 0.86, 0.8, 0.74, 0.68, 0.62, 0.56\
               , 0.5]) - 1.
lnIMF = np.array([6.63, 7.1, 7.36, 7.52, 7.72, 8., 7.98, 7.98, 8.32, 8.5, 8.6, 8.7, 8.83, 8.97, 9.04, \
                 9.13, 9.2, 9.22]) - 10.
```

```python
print np.std(lnIMF)
```

```python
def salpeter(theta, M):
    return theta[0] * M**(-theta[1])

def fun(theta, M, IMF):
    return salpeter(theta, M) - IMF
```

```python
theta0 = np.array([1., 1.])
IMF = 10**lnIMF
M = 10**lnM

res = least_squares(fun, theta0, jac='3-point', bounds=(0, 10), args=(M, IMF), verbose=1)
```

```python
res.x
```

```python
x = np.linspace(-0.4,1., 100)
y = salpeter(res.x, 10**x)
y2 = salpeter([0.03, 1.35], 10**x)

fig = plt.figure(figsize=(10,8))
plt.scatter(lnM, lnIMF)
plt.plot(x, np.log10(y), c='k', ls='--')
plt.plot(x, np.log10(y2), c='r', ls='-.')
plt.show()
```

## Part (b)

Now we're going to use `emcee` to fit for the value of $\alpha$ from the Salpeter data. But I'm very confused as to how this should be done, because we don't have actual mass sample to be able to fit a power law as for Problem 2. So I'm just going to go with trying to use `emcee` to fit a line to the data and hope for the best!

```python
### probabilites for emcee here
def lnlike(theta, x, y):
    '''We don't have error bars for Salpeter's data, so I'm just going to fit for those too!
    x = lnM
    y = lnIMF'''
    m, b, err = theta
    model = m * x + b
    inv_sigma2 = 1.0/(err**2)
    return -0.5*(np.sum((y-model)**2*inv_sigma2 - np.log(inv_sigma2)))

def lnprior(theta):
    m, b, err = theta
    if -5.0 < m < -1.0 and 0.0 < b < 10.0 and -5.0 < err < 5.0:
        return 0.0
    return -np.inf

def lnprob(theta, x, y):
    lp = lnprior(theta)
    if not np.isfinite(lp):
        return -np.inf
    return lp + lnlike(theta, x, y)
```

```python
### let's run emcee
ndim, nwalkers = 3, 100
pos = [np.array([-1.35, 0.03, 0.02]) + 1e-4*np.random.randn(ndim) for i in range(nwalkers)]
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob, args=(lnM, lnIMF))
sampler.run_mcmc(pos, 1000);
```

```python
[plt.plot(sampler.chain[i,:,0]) for i in range(100)]
#[plt.plot(sampler.chain[i,:,0]) for i in range(sampler.chain.shape[0])]
plt.xlabel('step')
plt.ylabel(r'm')
plt.show()

[plt.plot(sampler.chain[i,:,1]) for i in range(100)]
#[plt.plot(sampler.chain[i,:,0]) for i in range(sampler.chain.shape[0])]
plt.xlabel('step')
plt.ylabel(r'b')
plt.show()
```

```python
samples = sampler.chain[:, 250:, :].reshape((-1, ndim))
fig = corner.corner(samples, labels=["$m$", "$b$", "error"], show_titles=True)
plt.show(fig)
```
